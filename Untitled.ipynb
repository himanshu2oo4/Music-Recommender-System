```python
# Text preprocessing 

df['text'] = df['text'].str.lower().replace(r'^\w\s',' ',regex =True).replace(r'\r','', regex = True ).replace(r'\n','', regex = True )
```

```python
# removing the extra spaces from the text ---------

def joiner(text):
    return ' '.join(text.split())

df['text'] = df['text'].apply(joiner)
```

```python
df['text'][0]
```

```python
import nltk 
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
```

```python
def stemm(text):
    ls = []
    for i in text.split():
        ls.append(stemmer.stem(i))
        
    return ' '.join(ls)
```

```python
stemm('hello you are beautiful')

```

```python
df.shape
```

```python
df['text']  = df['text'].apply(stemm)
```

```python
df['text'][0]
```

```python
# removing stopwords while stemming 
from nltk.corpus import stopwords
word = stopwords.words('english')

def stop(text):
    lss = []
    for i in text.split():
        if i not in set(word):
            lss.append(i)
    
    return ' '.join(lss)
            
```

```python
df['text'] = df['text'].apply(stop)
```

```python
df['text'][0]
```

```python
def remove_duplicates(text):
    uq = []
    for i in text.split():
        if i not in uq :
            uq.append(i)
        
    return ' '.join(uq)

```

```python
df['text'] = df['text'].apply(remove_duplicates)
```

```python
df['text'][0]
```

```python
# ----- checking all the functions --------------

y = 'you are beautiful beauty' 

y = stemm(y)
print(y)
y = stop(y)
print(y)
y = remove_duplicates(y)
print(y)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfid = TfidfVectorizer(analyzer ='word',stop_words = 'english')
```

```python
matrix = tfid.fit_transform(df['text'])
```

```python
for i in matrix:
    print(i)
```

```python
similarity = cosine_similarity(matrix)
similarity 
```

```python
similarity.shape
```

```python
df[df['song'] == 'Waterfall'].index[0]
```

```python
similarity[0]
```

```python
# ----------- recommender func -----------

def recommend(music):
    idx = df[df['song'] == music].index[0]
    distance = sorted(list(enumerate(similarity[idx])), reverse = True , key = lambda x : x[1])
    song = []
    for sid in distance[1:11]:
        song.append(df.iloc[sid[0]].song)
    return song 
```

```python
recommend('Waterfall')   # recommended songs 
```
